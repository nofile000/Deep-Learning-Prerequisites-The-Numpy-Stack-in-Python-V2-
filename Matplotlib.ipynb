### StackPlot

```python
# distribution over a entire series of time
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [1,1,1,2,2,3,3,3,4]
player2 = [1,1, 2, 2,3,3,4,4,4]
player3 = [1,1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
```

```python
plt.stackplot(minutes,player1,player2,player3);
```

```python
# tracking whose working on the project the most
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
```

### Scatter Plot

```python
x = [7, 0, 3, 0, 5, 4, 8, 2, 3, 9]
y =  [1, 4, 0, 10, 1, 3, 1, 3, 9, 6]
plt.scatter(x,y)
```

```python
# setting size ,color , linewidth ,edgecolor marker
plt.scatter(x,y,s=100,c='red',linewidth =1,marker='X',alpha = 0.75)
```

```python
# taking reviews of people and plotting
ratings = [0, 10, 9, 2, 8, 8, 0, 4, 4, 4]
plt.scatter(x,y,s=100,c=ratings,cmap='Greens',linewidth=1,alpha = 0.72)
#plt.grid(True)
# lets plot the color bar to find indencity
cbar = plt.colorbar()
cbar.set_label('Satisfaction');
```

```python
# lets sized determine the ratings
ratings=[random.randint(0,100) for i in range(10)]
plt.scatter(x,y,s=ratings,linewidth = 5,alpha =0.75)
#plt.grid(False)
plt.show()

```

```python
# reading youtube ratings and views
#print(plt.style.available)
plt.style.use('seaborn')
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')
#print(data)
likes = data['likes']
ratio = data['ratio']
view_count = data['view_count']
plt.scatter(likes,view_count,s=10,c=ratio,cmap='summer',linewidth = 5,alpha=.72)
plt.xlabel('likes')
plt.ylabel('view_count')
plt.title('Youtube Views')
```

```python
# lets make it little more visible and plot the color bar
plt.scatter(likes,view_count,s=10,c=ratio,cmap='summer',linewidth = 1,alpha=.50)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('likes')
plt.ylabel('view_count')
plt.title('Youtube Views')
# plotting the color bar
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```
